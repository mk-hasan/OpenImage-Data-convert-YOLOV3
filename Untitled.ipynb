```python
class_desc =  pd.read_csv("/home/jupyter/yoloAtData/yolo/openimages/csv/class-descriptions-boxable.csv", names=["label","className"])
```

```python
class_dic =class_desc.set_index('label').T.to_dict('list')
```

```python
new_class_dic = {}
new_class_list =[]
```

```python
main_label = data_sorted["newLabelMap"]
```

```python
main_label = main_label.tolist()
```

```python
main_label
```

```python
for item in main_label:
    for key,val in class_dic.items():
        if(item == key):
            new_class_list.append(val)
            
        else:
            pass
                
            
```

```python
new_list_class = [item for sublist in new_class_list for item in sublist]
```

```python
data_sorted["className"] = new_list_class
```

```python
data_sorted.tail(100)
```

```python
class_order = data_sorted["className"].unique()
```

```python
class_order
```

```python
import os
for item in class_order:

    filename = "/home/jupyter/yoloAtData/yolo/openimages/csv/"+"oiclass.names"
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    with open(filename, "a") as f:
        val = item
        f.write(val)
        f.write("\n")
    f.close()
```
